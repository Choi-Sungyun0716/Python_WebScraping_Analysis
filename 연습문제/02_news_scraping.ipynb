```python
import requests
from bs4 import BeautifulSoup

#section_dict = {100:'정치',101:'경제',102:'사회',103:'생활/문화',104:'세계',105:'IT/과학'}

section_dict = {'기후/환경':'climate','사회':'society','경제':'economy','정치':'politics',
                '국제':'world','문화':'culture','생활':'life','IT/과학':'tech','인물':'people'}




def print_news(section):  #print_new(103,'생활/문화') 
    sid = section_dict[section]
    url = f'https://news.daum.net/{sid}'
    print(f'======> {url} {section} 뉴스 <======')

    req_header = {
    "user-agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36"
    
    }

    res = requests.get(url, headers=req_header)
    res.encoding = 'utf-8'
   

    if(res.ok):
        html = res.text 
        #print(html)
        # BeautifulSoup 객체 생성  
        soup = BeautifulSoup(html, "html.parser")
        #print(soup)
        # CSS 선택자
        a_tags = soup.select("ul.list_newsheadline2 li ") 
        



        for li_tag in a_tags:
            if(li_tag.text):
                a_tag = li_tag.find('a')
                link = a_tag["href"]

                strong_tag = li_tag.select_one('div.cont_thumb strong.tit_txt')
                title = strong_tag.text
                
                print(link)
                print(title)
            
# 응답(response)이 Error 이면 status code 출력 
    else:
      print(f"Error Code = {res.status_code}")
```

```python
print_news('경제')
print_news('사회')
```
